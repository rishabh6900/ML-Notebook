```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
df.sample(5)
```

```python
X = tips[['total_bill', 'size']]
y = tips['tip']
```

```python
df = df.iloc[:,2:]
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X,y,
                                                    test_size=0.3,
                                                    random_state=0)

X_train1.shape, X_test1.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python

from sklearn.preprocessing import StandardScaler

scaler1 = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler1.fit(X_train1)

# transform train and test sets
X_train_scaled1 = scaler1.transform(X_train1)
X_test_scaled1 = scaler1.transform(X_test1)
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```


Effect of Scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train1['total_bill'], X_train1['size'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled1['total_bill'], X_train_scaled1['size'],color='green')
ax2.set_title("After Scaling")
plt.show()
```

```python

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train1['total_bill'], ax=ax1)
sns.kdeplot(X_train1['size'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled1['total_bill'], ax=ax2)
sns.kdeplot(X_train_scaled1['size'], ax=ax2)
plt.show()
```


Why scaling is important?

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
lr1 = LinearRegression()
lr_scaled1 = LinearRegression()
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
# Fit the regression model
lr1.fit(X_train1, y_train1)
lr_scaled1.fit(X_train_scaled1, y_train1)
```

```python

lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred1 = lr1.predict(X_test1)
y_pred_scaled1 = lr_scaled1.predict(X_test_scaled1)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test1, y_pred1)
print("R-Squared (Actual):", r2)

r2_scaled = r2_score(y_test1, y_pred_scaled1)
print("R-Squared (Scaled):", r2_scaled)
```

```python

print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
dt_scaled.fit(X_train_scaled,y_train)
```

```python

y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
```

```python
df = pd.DataFrame({'Age': [25, 35, 45],
                   'EstimatedSalary': [50000, 150000, 200000],
                   'Purchased': [0, 1, 0]})

# New rows to be added
new_data = pd.DataFrame({'Age': [5, 90, 95],
                         'EstimatedSalary': [1000, 250000, 350000],
                         'Purchased': [0, 1, 1]})

# Use pd.concat to add rows
df = pd.concat([df, new_data], ignore_index=True)

print(df)
```

```python
df
```

```python
plt.scatter(df['Age'], df['EstimatedSalary'])
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```
