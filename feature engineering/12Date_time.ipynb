```python
#converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#Extract years
date['date_years'] = date['date'].dt.year
date.head()
```

```python
#Extract Months 
date['date_Months'] = date['date'].dt.month
date.head()
```

```python
#Extract Day
date['date_days'] = date['date'].dt.day
date.head()
```

```python
#Extract Months name
date['day_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# Day of week 
date['Day_of_week'] = date['date'].dt.dayofweek
date.head()
```

```python
# Day of year
date['Day_of_years'] = date['date'].dt.dayofyear
date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#Is weekwnd 
date['date_is_weekend '] = np.where(date['date_dow_name'].isin(['sunday','saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
# Extract Quarter
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract Semester

date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```
